```
import pandas as pd
import seaborn as sns

campaign = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/bank.csv')
campaign.head()
```

```
campaign['y'].value_counts()
```

```
campaign_plot = campaign[campaign["y"]== "yes"]
campaign_plot.head()
```

```
campaign["y_numeric"] = campaign["y"].map({"yes": 1, "no": 0})
```

```
response_rate= (campaign.groupby("previous")["y_numeric"].mean().reset_index())
response_rate.head()
```

```
sns.lineplot (data= response_rate, x = "previous", y = "y_numeric")
```
